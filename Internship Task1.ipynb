```python
#Q2- Python If-Else

n = int(input().strip()) #or: n=int(input("Enter a number:"))
if not(1<=n<=100): 
    print("Error: n must be between 1 and 100(0 is not allowed)")
else:
    if n%2!=0:
        print("Weird")  
    elif n%2 == 0 and 2<=n<=5:
        print("Not Weird")
    elif n%2 == 0 and 6<=n<=20:
        print("Weird")
    else: 
        print("Not Weird")   

```

```python
#Q6- Write a Function
def is_leap(year):
    leap = False
    if 1900 <= year <= 10**5:
        if year % 400 == 0:
            leap = True
        elif year % 100 == 0:
            leap = False
        elif year % 4 == 0:
            leap = True
    return leap
```
