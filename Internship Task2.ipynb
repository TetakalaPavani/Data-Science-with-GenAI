```python
#Q1-Running sum of 1d array

nums = list(map(int, input("Enter numbers: ").split()))            #or: import ast,# Take input as a list
running_sum = []                                                   #nums = ast.literal_eval(input("Enter a list, [1,2,3,4]:"))
total = 0         
for num in nums:
    total += num
    running_sum.append(total)
print(running_sum)  
```

```python
#Q2- Shuffle the array

import ast
n=int(input("Enter n (so that list length = 2n):"))
nums = ast.literal_eval(input("Enter a list:"))
result = []
leng=len(nums)
if 2*n == leng:  
    for i in range(n):
        result.append(nums[i])       
        result.append(nums[i+n])   
    print(result)
else:
    print(f"Invalid n! Your list has {leng}.")
    print("Enter n such that 2n equals the list length.")
```

```python
#Q3- Kids with the Greatest number of candies

import ast
candies = ast.literal_eval(input("Enter a list:"))
extraCandies = 3

n = len(candies)
result = [False] * n   

max_candy = max(candies)  
for i in range(n):
    new_candy = candies[i] + extraCandies
    result[i] = new_candy >= max_candy  
print(result)
```
